```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'html.parser')
table = soup.find('table',{'class':'wikitable sortable'})
headers="Postcode,Borough,Neighbourhood"
tableAux=""
for tr in table.find_all('tr'):
    row1=""
    for tds in tr.find_all('td'):
        row1=row1+","+tds.text
    tableAux=tableAux+row1[1:]
file=open("AssessmentQ2.csv","wb")
file.write(bytes(tableAux,encoding="ascii",errors="ignore"))
file.close()
```

```python
dfGeo = pd.read_csv('AssessmentQ2.csv')
dfGeo.columns=['Postalcode','Latitude','Longitude']
dfToronto = pd.merge(dfAux, dfGeo[['Postalcode','Latitude', 'Longitude']],on='Postalcode')
dfToronto
```
